```python
def returnRoofTypeDataFrame(place):

    tampaFlatt = data.groupby(["town"]).get_group(place)["roof_total_est_flat_roofs"].mean()
    tampaW = data.groupby(["town"]).get_group(place)["roof_total_est_w_facing"].mean()
    tampaN = data.groupby(["town"]).get_group(place)["roof_total_est_n_facing"].mean()
    tampaE = data.groupby(["town"]).get_group(place)["roof_total_est_e_facing"].mean()
    tampaS = data.groupby(["town"]).get_group(place)["roof_total_est_s_facing"].mean()

    return pd.DataFrame({"Flatt":tampaFlatt,
                                    "W":tampaW,
                                    "N":tampaN,
                                    "E":tampaE,
                                    "S":tampaS
                                      }, index=[0])
    return pd.DataFrame({'group':[],
         'day':['Mon','Tues','Fri','Thurs','Sat','Sun','Weds','Fri','Sun','Thurs','Sat','Weds','Mon','Tues'],
         'amount':[1,2,4,2,1,1,2,4,5,3,4,2,1,3]})


tampa = returnRoofTypeDataFrame("tamba")
newYork = returnRoofTypeDataFrame("new york city")
seattle = returnRoofTypeDataFrame("seattle")
albany = returnRoofTypeDataFrame("albany")







plot = newYork.plot(kind="bar",
         figsize=(30,20),
         fontsize=30,
         title="current output from roof type"
         )
plot.set_xlabel("roof type")
plot.set_ylabel("number megaWatts")
```
