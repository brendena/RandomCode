 
# cleaning the data 
 
### 1 
 
First i convert all the  unknowns to null, which makes filling in meaning full￼data a lot easier
 
### 2 
 
Second i delete all the column that don't have enough data to be meaning full.  
So if there is more then 90 % missing data i remove the column. 
 
### 3 
Third i delete all the row that don't have any unique values.  Some people 
suggested that for the column that where all True then that means that the 
rest must be false, but since this data was gathered through many different  
sources i didn't think i could make that conclusion. 
 
 
### 4 
Then to fill in the data i move a data point with most of the data to the top, because Interpolation needs a starting point. 
 
### 5 
 
FillMissingData function is a semi big function that calls 
specific function to fill in specific data types. 
 
if(Str): 
>    If its a string i treat it as catigorical data. 
>    This includes True or False values. 
>    To fill in i get the frequency of each value  
>    and times that ratio by the number of NaN  
>    elements.  Then i randomly put all the data 
>    in. 
 
if(int): 
>    If its a int i do Linear interpolation and 
>    then fill in any missing data with sum and  
>    round them to the nearest digit. 
 
if(date) -y-m-d 
>    If its a date i convert the date to the number of seconds 
>    from the 1900 january, 1. Then i do the same as i did with 
>    int and then i convert it back to a string.
